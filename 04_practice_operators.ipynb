```python
# ARITHMETIC OPERATORS

a = 10
b = 3

print(a + b) # Addition
print(a - b) # Subtraction
print(a * b) # Multiplication
print(a / b) # Division
print(a % b) # Modular (gives the remainder after division)
print(a ** b) # Exponentiation (power)
print(a // b) # Floor Division (to get in integer, rounds dows the value)
```

```python
# RELATIONAL (COMPARISON) OPERATORS
a = 15
b = 10

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
# AUGMENTED/ ASSIGNMENT OPERATORS

x = 5     # assignment operator
print(x)

x = 5     # assignment operator
x += 5    # augmented assignment (addition)
print(x)

x = 5     # assignment operator
x -= 5    # augmented assignment (subtraction)
print(x)

x = 5     # assignment operator
x *= 2    # augmented assignment (multiplication)
print(x)

x = 5     # assignment operator
x /= 4    # augmented assignment (division)
print(x)

x = 5     # assignment operator
x //= 3   # augmented assignment (floor division)
print(x)

x = 5     # assignment operator
x %= 3    # augmented assignment (modulus)
print(x)

x = 5     # assignment operator
x **= 2   # augmented assignment (power)
print(x)
```

```python
# LOGICAL OPERATORS 

# and
user_name = "admin@gmail.com"
password = 123456

# If both the values matched the answer will be true otherwise false
print(user_name == "admin@gmail.com" and password == 123456)
```

```python
# or
user_name = "admin@gmail.com"
password = 123456

# If one of the values matched the answer will be true otherwise false
print(user_name == "admin@gmail.com" or password == 12345)
```

```python
# not
user_name = "admin@gmail.com"
password = 123456

# If answer is true not operator will reverse it
print(not user_name == "admin@gmail.com" and password == 123456)
```
