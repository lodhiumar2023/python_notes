```python
# 1. No arguments, No return value.
# A function that takes no input from the user (no arguments) and does not return any value; it only performs a task and prints the result.
# All the logic and statements are written inside the function, so it performs the task without taking input or returning a value.

def add(): # No parameter
  a = int(input("Enter first value: "))
  b = int(input("Enter second value: "))
  total = a + b
  print("Result:", total)

add() # No argument
```

```python
# 2. With arguments no return value.

def add(a, b): # with parameter
  total = a + b
  print("Result:", total)

add(5, 6) # (pass a argument.)
```

```python
# 2. With arguments no return value.

def add(a, b): # with parameter
  total = a + b
  print("Result:", total)

value_1 = int(input("Enter first value: "))
value_2 = int(input("Enter second value: "))

add(value_1, value_2) # (can take input from user.)
```

```python
# 3. No arguments with return

# When a function uses a return statement, it sends back a value that can be stored or used in other parts of the program.
# print() → only shows the result.
# return → returns a value that can be stored or used in other parts of the program.

def add():
  a = int(input("Enter first value: "))
  b = int(input("Enter second value: "))
  total = a + b
  return total

result = add()
print("Answer is:", result)
```

```python
# 4. With arguments with return

# A function that takes input values (arguments) and returns a result after processing them.
# When a function uses a return statement, it sends back a value that can be stored or used in other parts of the program.
# print() → only shows the result.
# return → returns a value that can be stored or used in other parts of the program.

def add(a, b): # With parameter.
  total = a + b
  return total

value_1 = int(input("Enter first value: "))
value_2 = int(input("Enter second value: "))

result = add(value_1, value_1) # With argument.
print("Result is:", result)
```

```python
# Create a function to find Even or Odd.
# No argument, No return

def even_odd_num():
  num = 9

  if num % 2 == 0:
    print(num, "is an Even Number")
  else:
    print(num, "is an Odd Number")

even_odd_num()
```

```python
# Create a function to find Even or Odd.
# With argument, No return

def even_odd_num(num):

  if num % 2 == 0:
    print(num, "is an Even Number")
  else:
    print(num, "is an Odd Number")


num = int(input("Enter a number: "))
even_odd_num(num)
```

```python
# Create a function to find Even or Odd.
# No argument, with return

def even_odd_num():

  if num % 2 == 0:
    return num, "is an Even Number"
  else:
    return num, "is an Odd Number"

num = int(input("Enter a number: "))
result = even_odd_num()
print(result)
```
