```python
# 1️⃣ Greeting Function
# Create a function `greet_user(name)` that takes a name as a parameter and prints:

# Example Output:
# Hello, Ahmad!


def greet_user(name):
    user_name = (f"Hello, {name}!")
    print(user_name)


my_name = input("Please enter your name: ")
greet_user(my_name)
```

```python
# 2️⃣ Square Printer
# Create a function `print_square(number)` that takes a number and prints its square.

# Example:
# Input: 4
# Output: 16

def print_square(number):
    num_square = number ** 2
    print("Answer is: ", num_square)


my_number = int(input("Please enter a number to print square: "))
print_square(my_number)
```

```python
# 3️⃣ Even or Odd Checker
# Create a function `check_even_odd(num)` that takes a number and prints:

# - "Even" if the number is even
# - "Odd" if the number is odd


def check_even_odd(num):
    if num % 2 == 0:
        print(f"{num}: is an Even Number.")
    else:
        print(f"{num}: is an Odd Number.")


my_number = int(input('Please enter a number to check "Even" or "Odd": '))
check_even_odd(my_number)
```

```python
# 4️⃣ Full Name Printer
# Create a function `full_name(first, last)` that takes first name and last name and prints:

# Example Output:
# Your full name is: Ali Khan

def full_name(first, last):
    name = (f"Your full name is: {first} {last}")
    print(name)


first_name = input("First Name: ")
last_name = input("Last Name: ")
full_name(first_name, last_name)
```
