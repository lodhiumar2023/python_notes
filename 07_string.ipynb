```python
# String Slicing
# course[start:stop:step]

course = "python course"
        # 0123456789
print(course[0])     # access element at index 0 (first character)
print(course[-1])    # access element from the end (last character)
print(course[0:4])   # slicing from index 0 to 3 (4 is excluded)
print(course[:4])    # slicing from start (0) to 3 (4 is excluded)
print(course[0:])    # slicing from index 0 to the end
print(course[1:-1])  # slicing from index 1 to second last character (-1 is excluded)
print(course[::-1])  # print string in reverse order. 

print(course[-5::-1])  # print string in reverse order. 
```

```python
# String Functions 
# https://www.w3schools.com/python/python_ref_string.asp

str_1 = "I am python programer"

print(str_1.endswith("amer")) # returns true if string ends with substring.
print(str_1.capitalize()) # capitalize first character.
print(str_1.upper()) # change in upper case all string. 
print(str_1.lower()) # change in lower case all string. 
print(str_1.title()) # make each alphabet in capital of sting.
print(str_1.find("python"))
print(str_1.replace("python", "Java"))
```
