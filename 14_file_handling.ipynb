```python
# open a file in same directory.

f = open("data.txt", mode = "r")

if f:
    print("File opened successfully")
    print(type(f)) # f is a file handler to perform input/ output operations 
```

```python
# open a file in same from another directory.

f = open("/home/lodhiumar2023/Desktop/data2.txt", mode="r")

if f:
    print("File opened successfully")
    print(type(f)) # f is a file handler to perform input/ output operations 
```

```python
# buffering 10 means: 10 bytes, and utf is Unicode Transformation Format.
f = open("data.txt", mode="r", buffering=10, encoding="utf-8")

print("File handler: ", f)

if f:
    print("File opened")
```

```python
# File opening and file closing:
# Note: if we do not close file after performing operations, python garbage collector will destroy file object
# and close the file automatically. 
# Note: Don't rely on garbage collector. 

# filed opening
f = open("data.txt", mode="r", buffering=30, encoding= "utf-8")
if f:
    print("File is opened")

# filed closing
f.close()
if f:
    print("Filed is closed")
```
