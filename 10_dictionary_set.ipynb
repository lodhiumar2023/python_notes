**Dictionary**

```python
# Dictionary

my_dict = {
    "name" : "Kashif",
    "subject" : "English",
    "passing_marks" : 40
}

print(my_dict)
```

```python
# Can stored new key value pair.

my_dict = {
    "name" : "Kashif",
    "subject" : "English",
    "passing_marks" : 40
}

my_dict["percentage"] = "40%"

print(my_dict)
```

```python
# We can also store list or tuple data type in dictionary.

# List Data
list_itms = {
    "fruits" : ["Banana", "Mango", "Orange"],
    "mlik shakes" : ["Mango Shake", "Banana Shake, Straberry Shake"]
    }
print(type(list_itms["fruits"]))
print(list_itms)


# Tuple Data
list_itms = {
    "fruits" : ("Banana", "Mango", "Orange"),
    "mlik shakes" : ("Mango Shake", "Banana Shake", "Straberry Shake")
    }

print(type(list_itms["fruits"]))
print(list_itms)
```

```python
# Accessing Values
# We use the key to access its value.

my_dict = {
    "name" : "Kashif",
    "subject" : "English",
    "passing_marks" : 40
}

print(my_dict["name"])
print(my_dict["subject"])
print(my_dict["passing_marks"])

# to access any key value .get() fuction returns it.
my_dict.get("name")
```

```python
# New key value can be changed.

my_dict = {
    "name" : "Kashif",
    "subject" : "English",
    "passing_marks" : 40
}

my_dict["name"] = "Nasir"
print(my_dict)
```

```python
# Data can be appended (a single string) in dictionary if key value is a list.
list_itms = {
    "fruits" : ["Banana", "Mango", "Orange"],
    "mlik shakes" : ["Mango Shake", "Banana Shake", "Straberry Shake"]
    }

more_itms = "Black Barry"
list_itms["fruits"].append(more_itms)
print(list_itms)


# Data can be extend (a list) in dictionary if key value is a list.
list_itms = {
    "fruits" : ["Banana", "Mango", "Orange"],
    "mlik shakes" : ["Mango Shake", "Banana Shake", "Straberry Shake"]
    }

more_itms = ["Black Barry", "Blue Barry"]
list_itms["fruits"].extend(more_itms)
print(list_itms)
```

```python
# We can store nested dictionary

result_sheet = {
    "name_student" : ["Kashif", "Asif", "Mehmood", "Jibran"],
    # In a key a new dictionary can be stored.
    "Marks" : {
        "English" : 100,
        "Maths" : 100,
        "Physics" : 100,
        "Biology": 100,
        "Sindhi" : 75
    }

}

print(result_sheet["Marks"]["English"])
result_sheet["Marks"]["English"] = 75
print(result_sheet["Marks"])
```

```python
# While using copy() method, the dictionary‚Äôs ID will be different, but the inner list‚Äôs ID will be the same.

list_items = {
    "fruits" : ["Banana", "Mango", "Orange"],
    "mlik shakes" : ["Mango Shake", "Banana Shake", "Straberry Shake"]
    }

list_items1 = list_items
print(list_items1)

list_items2 = list_items.copy()
print(list_items2)

print(id(list_items1))
print(id(list_items2))
```

**Set**

```python
# Set

set_1 = {1, 2, 1, 2} # repeated elements stored only once, so it resolved {1, 2}
print(set_1, type(set_1))
```

```python
# We can create an empty set.

empty_set = set() # make empty set use set().
print(empty_set)

student_name = "Kashif" # This is an string
students_name = ("Sajid", "Asif", "Mehmood") # This is a tuple.

empty_set.add(student_name)
print(empty_set, type(empty_set))

empty_set.add(students_name)
print(empty_set, type(empty_set))
```
