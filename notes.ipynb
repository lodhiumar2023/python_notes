# Strings & Conditional Statement

**What is a String?**
>   1. String is a sequence of characters written inside quotes.
>   2. String can be written in Sindh, Double or Triple Quotes.\
>   
**Example:**
>   - Single quotes → 'Hello'
>   - Double quotes → "Hello"
>   - Triple quotes → '''Hello''' or """Hello""" (for multi-line text)

**String Indexing**
>   Each character has an index (starts from 0).

**Python String Methods (List with Short Comments)**

 **Case Changing Methods** 
>       upper()# Converts string to UPPERCASE \
>       lower()        # Converts string to lowercase \
>       title()        # First letter of each word capital \
>       capitalize()   # First letter capital \
>       swapcase()     # Upper ↔ Lower case change

**Checking Methods (True / False)** \
>       upper()# Converts string to UPPERCASE \
>       lower()        # Converts string to lowercase \
>       title()        # First letter of each word capital \
>       capitalize()   # First letter capital \
>       swapcase()     # Upper ↔ Lower case change

**Checking Methods (True / False)** \
> isupper()      # Checks if all letters are uppercase \
> islower()      # Checks if all letters are lowercase \
> istitle()      # Checks title case \
> isalpha()      # Only alphabets? \
> isdigit()      # Only digits? \
> isalnum()      # Alphabets + digits? \
> isspace()      # Only spaces?

**Search & Find Methods** \
> find()         # Finds position of character/word \
> rfind()        # Finds from right side \
> index()        # Like find(), but gives error if not found \
> count()        # Counts how many times value appears

**Replace & Modify Methods** \
> replace()      # Replaces old value with new value \
> strip()        # Removes spaces from both sides \
> lstrip()       # Removes spaces from left \
> rstrip()       # Removes spaces from right

**Split & Join Methods** \
> split()        # Breaks string into list \
> rsplit()       # Split from right \
> join()         # Joins list into string

**Starts & Ends Methods** \
> startswith()   # Checks starting word, character \
> endswith()     # Checks ending word, character

# Python Escape Sequence Characters (One-Line Description)

\n → Moves the cursor to a new line \
\t → Inserts a horizontal tab space \
\\ → Prints a backslash character \
\' → Allows a single quote inside a string \
\" → Allows a double quote inside a string \
\b → Removes the previous character (backspace) \
\r → Moves the cursor to the beginning of the line \
\f → Inserts a form feed (page break–like effect) \
\v → Inserts a vertical tab space \
\a → Triggers a system alert or beep sound \
\0 → Represents a null character \
\xhh → Represents a character using hexadecimal value \
\ooo → Represents a character using octal value

# Conditional Statements in Python

**Conditional statements are used to make decisions in a program.** 
> If condition is True → code runs \
> If condition is False → code is skipped

**Important Points**
- Python uses indentation (spaces) instead of { }
- Usually 4 spaces indentation is used
- Conditions use:
    - == equal
    - != not equal
    - '>'  greater than
    - < less than
    - '>=' greater or equal
    - <= less or equal

**Logical Operators in Conditions**
> and   # Both conditions True \
> or    # Any one True \
> not   # Reverse result

# Lists & Tuples

> - A built-in data type that stores set of values.
> - It can store elements of different types (integer, float, string, etc.).
> - List has indexing and slicing same as **String Indexing & Slicing.**
> - List is mutable (value can be change).
> - On print .appned() function do not return any statement. 

**List Method** 
> - **len()**   to find the length of a list
> - **list.append()** add only one element at the end, if needed to add, have to sent in form of object [1, 2, 3]
> - **list.extend()** is used to add multiple values in a list at once.
> - **list.sort()** sort by default in ascending order.
> - **list.sort(reverse=True)** to sort in descending order.
> - **list.reverse()** reverse the original list and do not sort.
> - **list.insert(index, element)** to adds a value on specific index.
> - **list.remove(1)** removes first occurrence of a element. from [1,1,2,3] to [1,2,3].
> - **list.pop(index)** removes element at index.
