# While Loop

‚úÖ Definition of While Loop

- A while loop is a control flow statement in Python that repeatedly executes a block of code as long as a given condition is True. 
- The loop stops automatically when the condition becomes False.

**Basic Syntax:**

```python
while condition:
    # code block

# If the condition is True, the code runs.
# If the condition becomes False, the loop stops.
```

‚úÖ Categories / Styles of While Loop.

A while loop can generally be written in four main styles:

1Ô∏è‚É£ Counter-Controlled While Loop \
2Ô∏è‚É£ Condition-Based While Loop \
3Ô∏è‚É£ Sentinel-Controlled While Loop \
4Ô∏è‚É£ Infinite While Loop (Using break)

```python
# 1. Counter-Controlled While Loop
#    This type of loop runs a fixed number of times using a counter variable.

count = 1

while count <= 5:
    print("Hello")
    count += 1

# When to Use:
# Printing numbers from 1 to 10
# Giving 3 or 5 fixed attempts
# Running a loop a specific number of times
```

```python
# 2. Condition-Based While Loop
#    This loop runs as long as a condition remains True.

num = int(input("Enter number: "))

while num > 0:
    print("Positive number")
    num = int(input("Enter number: "))

# When to Use:
# When working with user input
# When the number of repetitions is not fixed
# When execution depends on a condition
```

```python
# 3. Sentinel-Controlled While Loop
#    This loop runs until a special stop value (sentinel value) is entered.
#    A sentinel value is a specific value that signals the loop to stop.

num = int(input("Enter number (-1 to stop): "))

while num != -1:
    print("You entered:", num)
    num = int(input("Enter number (-1 to stop): "))

# When to Use:
# When the user can stop the program manually
# Data entry systems
# Input collection programs
```

```python
# 4. Infinite While Loop (Using break)
#    This loop runs forever unless stopped manually using the break statement.

while True:
    password = input("Enter password: ")
    
    if password == "python123":
        print("Login successful")
        break
    else:
        print("Wrong password")

# When to Use:
# Login systems
# Menu-driven programs
# Games
# Retry systems
```
