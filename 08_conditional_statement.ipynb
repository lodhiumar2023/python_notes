```python
# Question 1
#Student Grade System
    # 90-100 → A+
    # 80-89 → A
    # 70-79 → B
    #60-69 → C
    # Below 60 → Fail

obtained_marks = int(input("Please put your obtained marks: "))

if obtained_marks >= 90:
    print("A+")
elif obtained_marks >= 80:
    print("A")
elif obtained_marks >= 70:
    print("B")
elif obtained_marks >= 60:
    print("C")
else:
    print("Failed")
```

```python
# Electricity Bill Calculator
# Input:
# Units consumed
# Rules:
# 0-100 units → Rs 5 per unit
# 101-200 units → Rs 7 per unit
# 201-300 units → Rs 10 per unit
# Above 300 → Rs 15 per unit

unit_consumed = int(input("Please enter consumed units: "))

if unit_consumed <= 100:
    print("Your Bill is:",unit_consumed * 5)
elif unit_consumed >= 101 and unit_consumed <= 200:
    print("Your Bill is:", unit_consumed * 7)
elif unit_consumed >= 201 and unit_consumed <= 300:
    print("Your Bill is:", unit_consumed * 10)
elif unit_consumed > 300:
    print("Your Bill is:", unit_consumed * 15)
else: 
    print("Please enter consumed units")
print("Thank you for using Electricity Calculator")
```
