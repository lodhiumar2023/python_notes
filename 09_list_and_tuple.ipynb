```python
# Data type List

students_name = ["Kashif", "Nasir", "Umair", "Yasir"]   # indexing [0, 1, 2, 3]
print(students_name)

quiz_list = ["Cake", 10, "bottle", "Grapes", 44, 45, 48]    # indexing [0, 1, 2, 3,]
print(len(quiz_list))   # length of list can also check using len() function same as in string. 

# slicing
print(quiz_list[1:5:1])
print(quiz_list[1:3])
print(quiz_list[1::])
print(quiz_list[-1::-1])
```

```python
# list.append() function 

students_name = ["Kashif", "Nasir", "Umair", "Yasir"]   # indexing [0, 1, 2, 3]
students_name.append(["Noor", "Ahmed"]) # added as a one object, not two different values. 
print((students_name))
print((students_name[4]))
```

```python
# list.extend() function 
students_name = ["Kashif", "Nasir", "Umair", "Yasir"]
students_name.extend(["Noor", "Ahmed"])  # use to add multiple values at once. 
print(students_name)
```

```python
# list.sort()
students_name = ["Umair", "Nasir", "Yasir","Kashif"]
students_name.sort()
print(students_name)

# list.sort(reverse=True)
students_name = ["Umair", "Nasir", "Yasir","Kashif"]
students_name.sort(reverse=True)
print(students_name)
```

```python
# list.insert(Index, Value)

students_name = ["Umair", "Nasir", "Yasir","Kashif"]
students_name.insert(1,"Muhammad")
print(students_name)
```

```python
# list.remove(Fist Occurrence of a element)
list_serial = [1,2,1,3,4,5,6,]
list_serial.remove(1)
print(list_serial)
```

```python
# list.pop(Index)
list_serial = [1,2,1,3,4,5,6,]
list_serial.pop(1)
print(list_serial)
```
