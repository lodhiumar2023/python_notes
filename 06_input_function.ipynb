```python
# Write a program to height and width of a rectangle and its area. 

height = int(input("Please give height: "))
width = int(input("Please give width: "))

print("Height", height)
print("Width", width)


area = height * width
print("Area of rectangle is: ", area)
```

```python
# Write a program to take two inputs in float and print their average. 

value_1 = float(input("Write value 1: "))
value_2 = float(input("Write value 2: "))

print(value_1)
print(value_2)

average = (value_1 + value_2) / 2
print(average, type(average))
```

```python
# Write a program to take two inputs in integers and get answer in True or False (Boolean)

a = int(input("Value a: "))
b = int(input("Value b: "))

print("Value a: ", a)
print("Value b: ", b)

answer = a >= b
print("a is greater than and equal to b: ", answer)
```
