```python
# - Q1. Print numbers from 1 to 20 using a while loop.

counter = 1

while counter <= 5:
    print(counter)
    counter += 1
```

```python
# - Q2. Print only even numbers from 2 to 20 using a while loop.

counter = 1

while counter <= 20:
        if counter %2 == 0:
            print(counter)
        counter += 1
```

```python
# - Q3. Calculate and print the sum of numbers from 1 to 50 using a while loop.

num_list = []
counter = 1

while counter <= 5:
    num_list.append(counter)
    counter += 1

total = sum(num_list)
print("total of num_list is: ", total)
```

```python
# - Q4. Print the multiplication table of 7 using a while loop.

table = 2

counter = 1

while counter <= 10:
    print(table * counter)
    counter += 1
```

```python
# - Q5. Ask the user to enter a number.
#       Keep printing "Number is positive" while the number is greater than 0.


user_number = int(input("Please enter a positive number: "))

while user_number > 0:
    
    print(user_number)
    user_number = int(input("Please enter a positive number: "))
print("Loop ended")
```

```python
# Q6. - Ask the user to enter marks.
#       Keep asking while marks are less than 40.
#       When marks become 40 or more, print "You Passed".

user_marks = int(input("Please enter marks: "))

while user_marks < 40:
    print("You failed. Marks must be at least 40 to pass.")
    user_marks = int(input("Please enter marks again: "))

print("You Passed!")
```

```python
# Q7. - Ask the user to enter a password.
#       Keep asking while the password length is less than 6 characters.

user_password = input("please enter password: ")

while len(user_password) < 6:
    print("Password must be at least 6 characters long!")
    user_password = input("please enter password: ")

print("You login successful!")
```

```python
# Q8. - Keep asking the user to enter numbers.
#       Stop the loop when the user enters 0.
#       Print the total sum of all entered numbers (except 0).


total = 0
user_number = int(input("Please enter a number: "))

while user_number != 0:
    total += user_number
    user_number = int(input("Enter a number or 0 to stop loop: "))

print("Total sum:", total)
```

```python
# Q9. - Ask the user to enter names.
#       Stop when the user enters "exit".
#       Print all names entered (you may store them in a list).


user_input = input("Please enter name: ")
user_data = []

while user_input.lower() != "exit":
    user_data.append(user_input)
    user_input = input("Enter exit to close menu:")

print(user_data)
```

```python
# Q10. - Keep taking input from the user until they enter -1.
#        Count how many numbers were entered (excluding -1).

user_input = int(input("Please enter a number: "))
user_data = []

while user_input != -1:
    user_data.append(user_input)
    user_input = int(input("Please enter a number or -1 to stop: "))

print("Total numbers entered:", len(user_data))
```

```python
# Q11. - Create a simple login system:
#        Correct userName = "admin"
#        Correct userPassword = "1234"
#        Keep asking until both are correct.


userName = "admin"
userPassword = "1234"
counter = 0

while True:

    input_userName = input("Enter user name: ")
    input_userPassword = input("Enter user password: ")

    counter += 1

    if input_userName == userName and input_userPassword == userPassword:
        print("Login Successful!")
        break
    else:
        print("Invalid Credentials")

    if counter == 3:
        print("Your attempted 3 times")
        break
```

```python
# Q12. - Create a simple menu:
# Add
# Subtract
# Exit
# Keep showing the menu until the user chooses Exit.

menu_list = ["Add", "Subtract"]

while True:
    userInput = input("Enter one: Add, Subtract or Exit to close: ")

    if userInput == "Exit":
        print("Thanks for using")
        break

    elif userInput not in menu_list:
        print("You chose wrong menu")

    elif userInput in menu_list:
        print(userInput)
```

```python
# Q13. - Keep asking the user to guess a number (correct number = 5).
#        Stop the loop when the guess is correct.


while True:
    inputNumber = int(input("Please guess a number:"))

    if inputNumber == 5:
        print("You guess right number", inputNumber)
        break
```

```python
# Q14.
# - Ask the user to enter numbers continuously.
# - Stop when they enter "stop".
# - At the end, print:

# Total numbers entered \       length of list
# Maximum Number \              list.max()
# Minimum Number                list.min()


userData = []

while True:
    userInput = input("Please enter a number or stop to exit: ")

    if userInput == "stop":
        print("Thanks for using")
        break
    else:
        userData.append(int(userInput))

print("Total enter numbers are:", len(userData))
print("Sum of number is:", sum(userData))
print("Max of number is:", max(userData))
print("Min of number is:", min(userData))
```
